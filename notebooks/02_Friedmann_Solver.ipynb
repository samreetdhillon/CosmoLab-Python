### Solving the Fate of the Universe
By integrating the Friedmann equation, we can see how the scale factor $a(t)$ behaves.
- If $a(t)$ reaches 0 again, it's a Big Crunch.
- If $a(t)$ grows linearly, it's an Empty Universe.
- If $a(t)$ grows exponentially, it's $\Lambda$-dominated.

```python
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
from src.physics_engine import CosmologyEngine
from src.solvers import solve_scale_factor

# Define different cosmological scenarios
models = {
    "Planck 2018 (Flat)": CosmologyEngine(H0=67.4, Om0=0.31, Ol0=0.69),
    "Matter Only (SCDM)": CosmologyEngine(H0=70, Om0=1.0, Ol0=0.0),
    "Closed (Big Crunch)": CosmologyEngine(H0=70, Om0=2.0, Ol0=0.0),
    "Empty Universe": CosmologyEngine(H0=70, Om0=0.0, Ol0=0.0)
}

plt.figure(figsize=(10, 6))
t_span = (0, 30) # 0 to 30 Billion Years

for name, cosmo in models.items():
    t, a = solve_scale_factor(t_span, cosmo)
    plt.plot(t, a, label=name)

plt.axhline(1, color='black', linestyle='--', alpha=0.5, label="Today (a=1)")
plt.xlabel("Time since Big Bang (Gyr)")
plt.ylabel("Scale Factor $a(t)$")
plt.title("The Evolution of the Scale Factor")
plt.ylim(0, 3)
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
