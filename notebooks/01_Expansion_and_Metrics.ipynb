### The FLRW Metric and the Scale Factor
The metric for a homogeneous, isotropic universe is:
$$ds^2 = -c^2 dt^2 + a(t)^2 \left[ \frac{dr^2}{1-kr^2} + r^2(d\theta^2 + \sin^2\theta d\phi^2) \right]$$
Here, $a(t)$ is the **scale factor**, representing how distances "stretch" over time.

```python
import sys
import os
sys.path.append('../') # Access the src folder

import numpy as np
import matplotlib.pyplot as plt
from src.physics_engine import CosmologyEngine

# Initialize a Flat Lambda-CDM model
cosmo = CosmologyEngine(H0=67.4, Om0=0.315, Ol0=0.685)

# Generate a range of redshifts
z_range = np.linspace(0, 5, 100)
h_z = cosmo.hubble_parameter(z_range)

plt.figure(figsize=(8, 5))
plt.plot(z_range, h_z, label="$\Lambda$CDM", color='darkblue', lw=2)
plt.xlabel("Redshift (z)")
plt.ylabel("Hubble Parameter H(z) [km/s/Mpc]")
plt.title("The Expansion History of the Universe")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
