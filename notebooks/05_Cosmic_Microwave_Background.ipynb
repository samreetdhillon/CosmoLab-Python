## The CMB and the Surface of Last Scattering
Before $z \approx 1100$, the universe was a hot, ionized plasma. Photons were trapped, constantly scattering off free electrons. Once the universe cooled enough for neutral hydrogen to form (Recombination), photons became free to travel.

We analyze these fluctuations using the Angular Power Spectrum ($C_\ell$). The position of the "peaks" tells us the geometry of the universe.

```python
import numpy as np
import matplotlib.pyplot as plt

# In a real scenario, you'd use a library like CAMB or CLASS.
# Here, we create a pedagogical model of the peaks.
def mock_cmb_power_spectrum(ell):
    """Generates a simplified version of the CMB peaks."""
    # First peak (Fundamental), Second (Damping), etc.
    peak1 = 5000 * np.exp(-(ell - 220)**2 / (2 * 50**2))
    peak2 = 2500 * np.exp(-(ell - 540)**2 / (2 * 40**2))
    peak3 = 1500 * np.exp(-(ell - 800)**2 / (2 * 40**2))
    noise = 500 * np.exp(-ell/1000)
    return peak1 + peak2 + peak3 + noise

ell = np.arange(2, 1200)
cl = mock_cmb_power_spectrum(ell)

plt.figure(figsize=(10, 6))
plt.plot(ell, cl, color='darkorange', lw=2)
plt.fill_between(ell, cl, color='orange', alpha=0.1)

# Highlight the First Peak
plt.annotate('First Peak: Flatness', xy=(220, 5000), xytext=(350, 5500),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel("Multipole Moment ($\ell$)")
plt.ylabel("Temperature Variance $\ell(\ell+1)C_\ell / 2\pi$")
plt.title("The CMB Angular Power Spectrum")
plt.grid(alpha=0.3)
plt.show()
```
